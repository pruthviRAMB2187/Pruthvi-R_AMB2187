```python
from __future__ import print_function
# 1.IF-ELIF-ELSE (more than two options)
print("\n=== IF-ELIF-ELSE ===")
soil_colour = "Red"

if soil_colour in "Red":
    print("soil is more fertile")
    print("possibility of more yield.")
elif soil_colour == "black":
    print("soil is little fertile and clayey")
    print("possibility of little yield")
else:
    print("soil is not good for agriculture")
```
